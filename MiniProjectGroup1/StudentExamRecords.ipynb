```python
!wget -O reading_writing.csv https://raw.githubusercontent.com/xeweva/CS-131/main/MiniProjectGroup1/data/Shervan/reading_writing_scores.csv
```

```python
df2 = pd.read_csv("reading_writing.csv", encoding='latin-1')
```

```python
df2.plot.scatter(x= 'reading score', y='writing score', s=8)
plt.title('Correlation Between Reading and Writing Scores')

x = df2['reading score']
y = df2['writing score']

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x), "r--")
```

```python
education_vs_scores = pd.read_csv('https://raw.githubusercontent.com/xeweva/CS-131/main/MiniProjectGroup1/data/Bhavdeep/parental_education_scores.csv')
import seaborn as sns

plt.figure(figsize=(10, 6))
blue_palette = sns.color_palette("Blues", len(education_vs_scores))
plt.bar(education_vs_scores['Parental Level of Education'], education_vs_scores['Average Test Score'], color=blue_palette)
plt.title('Avg Test Scores vs Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Test Score')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
!wget -O lunchMeans.txt https://raw.githubusercontent.com/xeweva/CS-131/main/MiniProjectGroup1/data/Max/lunchMeans.txt
```

```python
# Reading data from the file
with open('lunchMeans.txt', 'r') as file:
    lines = file.readlines()

# Parsing categories and mean values
categories = []
mean_values = []
for line in lines:
    category, mean = line.split(':')
    categories.append(category.strip())
    mean_values.append(float(mean.strip()))

# Generating a range for the number of categories
y_pos = np.arange(len(categories))

# Creating the bar plot
plt.bar(y_pos, mean_values, align='center', alpha=0.7)
plt.xticks(y_pos, categories)
plt.ylabel('Average Score')
plt.title('Lunch Type')

# Displaying the plot
plt.show()
```
